### Missing Values 


```python
features_with_na = [feature for feature in dataset.columns if dataset[feature].isnull().sum()]
```

```python
for feature in features_with_na:
    print(feature,np.round(dataset[feature].isnull().mean()),4)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Loop through features with missing values
for feature in features_with_na:
    # Create a binary indicator: 1 if missing, 0 if not
    data[feature + '_na'] = np.where(data[feature].isnull(), 1, 0)

    # Plot the median SalePrice for missing vs. non-missing values
    data.groupby(feature + '_na')['SalePrice'].median().plot.bar()
    plt.title(f'{feature} - Missing vs Non-Missing')
    plt.xlabel('Missing (1) vs Not Missing (0)')
    plt.ylabel('Median SalePrice')
    plt.show()
```

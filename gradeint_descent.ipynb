```python
import numpy as np

class LinearRegressionGD:
    def __init__(self, learning_rate=0.08, iterations=10000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.m = 0
        self.b = 0

    def fit(self, x, y):
        n = len(x)

        for i in range(self.iterations):
            y_predicted = self.m * x + self.b
            cost = (1/n) * sum((y - y_predicted) ** 2)
            md = -(2/n) * sum(x * (y - y_predicted))
            bd = -(2/n) * sum(y - y_predicted)

            self.m -= self.learning_rate * md
            self.b -= self.learning_rate * bd

            if i % 100 == 0:  # Print every 100th iteration
                print(f"Iteration {i}: m = {self.m}, b = {self.b}, cost = {cost}")

        return self

    def predict(self, x):
        return self.m * x + self.b

# Example usage:
x = np.array([1, 2, 3, 4, 5])
y = np.array([5, 7, 9, 11, 13])

model = LinearRegressionGD()
model.fit(x, y)
predictions = model.predict(x)
print("Predictions:", predictions)
```
